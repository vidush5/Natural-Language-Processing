```python
# Keeping only relavant columns
df_final = df_final[['movie_id','title','overview','genres','keywords','cast','crew']]
df_final.head()
```

```python
# Checking the data contains any null values
df_final.isnull().sum()
```

```python
# Checking the data contains any duplicated records
df_final.duplicated().sum()
```

```python
# Dropping the null value records
df_final.dropna(inplace=True)
```

```python
df_final.shape
```

Earlier the data set size was (4809, 7) but we noticed that in the overview column 3 records were null so we removed them from our data set.

```python
df_final.head(2)
```

```python
def str_to_list(text):
    lists = [i['name'] for i in ast.literal_eval(text)]
    return lists
```

```python
df_final['genres'] = df_final['genres'].apply(str_to_list)
```

```python
df_final.head(2)
```

```python
df_final['keywords'] = df_final['keywords'].apply(str_to_list)
df_final.head(2)
```

```python
def convert_cast(text):
    counter = 0
    L = [i['name'] for i in ast.literal_eval(text) if counter < 3]
    counter+=1
    return L
```

```python
df_final['cast'] = df_final['cast'].apply(convert_cast)
df_final.head()
```

```python
def fetch_director(text):
    output = [i['name'] for i in ast.literal_eval(text) if i['job'] == 'Director']
    return output
```

```python
df_final['crew'] = df_final['crew'].apply(fetch_director)
df_final.head()
```

```python
# Split overview column
df_final['overview'] = df_final['overview'].apply(lambda x:x.split())
df_final.head()
```

```python
def remove_space(L):
    L1 = [i.replace(" ","") for i in L]
    return L1
```

```python
df_final['cast'] = df_final['cast'].apply(remove_space)
df_final['crew'] = df_final['crew'].apply(remove_space)
df_final['genres'] = df_final['genres'].apply(remove_space)
df_final['keywords'] = df_final['keywords'].apply(remove_space)

df_final.head()
```

```python
# Concatenate all columns
df_final['tags'] = df_final['overview'] + df_final['genres'] + df_final['keywords'] + df_final['cast'] + df_final['crew']
df_final.head()
```

```python
# droping unwanted columns
df_final = df_final[['movie_id','title','tags']]
df_final.head()
```

```python
# Converting list to str
df_final['tags'] = df_final['tags'].apply(lambda x: " ".join(x))
df_final.head()
```

```python
df_final['tags'] = df_final['tags'].apply(lambda x:x.lower())
df_final.head()
```

## Similarity Model Building

```python
import nltk
from nltk.stem import PorterStemmer
```

```python
ps = PorterStemmer()
```

```python
def stems(text):
    T = [ps.stem(i) for i in text.split()]
    return " ".join(T)
```

```python
df_final['tags'] = df_final['tags'].apply(stems)
df_final.head()
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=5000,stop_words='english')
```

```python
vector = cv.fit_transform(df_final['tags']).toarray()
vector[0]
```

```python
vector.shape
```

```python
len(cv.get_feature_names())
```

```python
from sklearn.metrics.pairwise import cosine_similarity
similarity = cosine_similarity(vector)
```

```python
similarity
```

```python
similarity.shape
```

```python
def recommend(movie):
    index = df_final[df_final['title'] == movie].index[0]
    distances = sorted(list(enumerate(similarity[index])),reverse=True,key = lambda x: x[1])
    
    recommeded_movies = [df_final.iloc[i[0]].title for i in distances[1:6]]
    
    return recommeded_movies
```

```python
recommend('Spider-Man 2')
```
